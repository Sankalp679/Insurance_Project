Remove all where claim_id, policy_id, claim_status, claim_amount, lastupdated, null

Remove all rows where policy_id not exists in policy table 

Convert date_of_claim to Date column with format (mm-dd-yyyy)

Ensure claim amount is >0 

```
# date_format returns string to convert into date to_date used so that date operations can be done on that column

answer_result = spark.sql("select claim.claim_id,claim.policy_id,to_date(date_format(claim.date_of_claim,'MM-dd-yyyy'),'MM-dd-yyyy') as date_of_claim,claim.claim_amount,claim.claim_status,claim.LastUpdatedTimeStamp from bronzelayer.claim inner join bronzelayer.policy on claim.policy_id = policy.policy_id where claim.claim_id is not null and claim.policy_id is not null and claim.claim_amount is not null and claim.LastUpdatedTimeStamp is not null and claim.merge_flag == false and claim.claim_amount>0")

display(answer_result)
```

Add the merged_date_timestamp (current timestamp)

```
answer_result.createOrReplaceTempView("clean_claim")

spark.sql("MERGE INTO silverlayer.claim as t using clean_claim as s on t.claim_id = s.claim_id when matched then update set t.policy_id = s.policy_id,t.date_of_claim = s.date_of_claim,t.claim_amount = s.claim_amount,t.claim_status = s.claim_status,t.LastUpdatedTimeStamp = s.LastUpdatedTimeStamp,t.merged_timestamp= current_timestamp() when not matched then insert (claim_id,policy_id,date_of_claim,claim_amount,claim_status,LastUpdatedTimeStamp,merged_timestamp) values (s.claim_id,s.policy_id,s.date_of_claim,s.claim_amount,s.claim_status,s.LastUpdatedTimeStamp,current_timestamp())")
```

Update the merged flag in the bronzelayer table

```
spark.sql("UPDATE bronzelayer.claim set merge_flag = true where merge_flag = false")

%sql select * from silverlayer.claim;
```
