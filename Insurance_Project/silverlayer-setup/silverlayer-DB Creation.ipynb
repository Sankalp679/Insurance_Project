```
%sql
create or replace table silverlayer.Agent
(
    agent_id integer,
    agent_name string,
    agent_email string,
    agent_phone string,
    branch_id integer,
    create_timestamp timestamp,
    merged_timestamp timestamp
) USING DELTA LOCATION "/mnt/silverlayer123/Agent"
```

```
%sql create or replace table silverlayer.Branch
(
branch_id int,
branch_country string,
branch_city string,
merged_timestamp timestamp
) USING DELTA LOCATION '/mnt/silverlayer123/Branch'
```

```
%sql create or replace table silverlayer.Customer
(
customer_id int,
first_name string,
last_name string,
email string,
phone string,
country string,
city string,
registration_date timestamp,
date_of_birth timestamp,
gender string,
merged_timestamp timestamp
) USING DELTA LOCATION '/mnt/silverlayer123/Customer'
```

```
%sql create or replace table silverlayer.Claim
(
claim_id int, 
policy_id int, 
date_of_claim date,
claim_amount double, 
claim_status string, 
LastUpdatedTimeStamp timestamp,
merged_timestamp timestamp 
) USING DELTA LOCATION '/mnt/silverlayer123/ClaimData'
```

```
%sql create or replace table silverlayer.Policy
(
policy_id int,
policy_type string,
customer_id int,
start_date timestamp,
end_date timestamp,
premium int,
coverage_amount int,
merged_timestamp timestamp 
) USING DELTA LOCATION '/mnt/silverlayer123/PolicyData'
```
